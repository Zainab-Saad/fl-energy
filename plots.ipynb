```python
import matplotlib.pyplot as plt

C = [0.0, 0.1, 0.2, 0.5, 1.0]
iid_b10  = [0.7612, 0.7607, 0.9151, 0.9104, 0.9134]
iid_binf = [0.2937, 0.2889, 0.2763, 0.2696, 0.2003]
non_b10  = [0.7128, 0.8929, 0.9061, 0.9055, 0.909]
non_binf = [0.2642, 0.2183, 0.2521, 0.2841, 0.2906]

plt.plot(C, iid_b10, 'o-', label='IID B=10')
plt.plot(C, iid_binf, 'o--', label='IID B=')
plt.plot(C, non_b10, 's-', label='Non-IID B=10')
plt.plot(C, non_binf, 's--', label='Non-IID B=')

baseline = 0.9263
plt.axhline(y=baseline, color='purple', linestyle='--', linewidth=1.2, label=f'Centralized Baseline ({baseline:.4f})')

plt.xlabel("Fraction of Clients (C)")
plt.ylabel("Test Accuracy")
plt.title("Effect of Client Fraction (C) on Test Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```
